```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the function and its derivative (gradient)
def function(w):
    return (w - 3) ** 2  # Quadratic function

def gradient(w):
    return 2 * (w - 3)  # Derivative (Gradient)
```

```python
# Initialize parameters
w = np.random.randn()  # Random initial weight
learning_rate = 0.1  # Step size
epochs = 20  # Number of iterations
```

```python
# Store values for visualization
w_history = []
loss_history = []
```

```python
# Perform Gradient Descent
for i in range(epochs):
    grad = gradient(w)  # Compute gradient
    w = w - learning_rate * grad  # Update weight
    loss = function(w)  # Compute loss

    # Store for visualization
    w_history.append(w)
    loss_history.append(loss)

    print(f"Epoch {i+1}: w = {w:.4f}, Loss = {loss:.4f}")
```

```python
# Plot Weight Updates
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(range(epochs), w_history, marker='o', linestyle='-')
plt.xlabel("Epoch")
plt.ylabel("Weight (w)")
plt.title("Weight Updates")
```

```python
# Plot Loss Curve
plt.subplot(1,2,2)
plt.plot(range(epochs), loss_history, marker='o', linestyle='-', color='r')
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Loss Decrease")

plt.show()
```
